```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
#使用chrome的webdriver
chrome_options = Options() # 啟動無頭模式
chrome_options.add_argument('--headless')  #規避google bug
chrome_options.add_argument('--disable-gpu')
```

```python
def get_report_html(ticker,years,quaters):
    executable_path = './chromedriver_win32/chromedriver.exe'
    browser = webdriver.Chrome(executable_path=executable_path,options=chrome_options)
    browser.get('https://mops.twse.com.tw/server-java/t164sb01?step=1&CO_ID='+ticker+'&SYEAR='+years+'&SSEASON='+quaters+'&REPORT_ID=C')
    html = browser.page_source
    print("fetch success")
    browser.quit()
    return html
```

```python
html = get_report_html('2330','2019','2')
```

```python
import pandas as pd
from bs4 import BeautifulSoup
```
